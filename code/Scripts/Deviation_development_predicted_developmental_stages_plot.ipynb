Load paths from config file

```python
from twinnet_tools.tnconfig import ProjectConfig

config = ProjectConfig("twinnet_config")
```

```python
dir_root_scripts = config.json["dir_scripts"]
path_model_zebrafish = config.json["dir_model_zebrafish_original"]
paths_reference_src_images = config.json["path_json_reference_src_imgs_2"]

config_paths_script = config.json["Deviation_development_predicted_developmental_stages"]
dir_src_sims = f'{config_paths_script["dir_data_dst"]}/similarities'
dir_dst_plots = f'{config_paths_script["dir_data_dst"]}/plots'
```

Import Python packages

```python
import glob
import matplotlib as mpl
from pathlib import Path
import sys

sys.path.append(dir_root_scripts)
from twinnet_tools.tnarrowplot import TNToolsArrowPlot
```

Prepare class instances

```python
tools_arrow_plot  = TNToolsArrowPlot()
```

Adjust matplotlib parameters to save plots as .svg files.

```python
new_rc_params = {'text.usetex': False,
                'svg.fonttype': 'none'}
mpl.rcParams.update(new_rc_params)
```

Function to format outputs

```python
def info_util(string_info):
    print('\033[1m[INFO] \033[0m' + str(string_info))
```

## Load paths <a class="anchor" id="second-bullet"></a>

```python
dir_dst_plot_path = Path( dir_dst_plots )
dir_dst_plot_path.mkdir(parents=True, exist_ok=True)
```

Load paths of analyzed embryos

```python
paths_normal = glob.glob(f"{dir_src_sims}/paths_normal/*")
paths_disintegrating = glob.glob(f"{dir_src_sims}/paths_disintegrating/*")
paths_patterning_defects = glob.glob(f"{dir_src_sims}/paths_patterning_defects/*")
```

```python
info_util(f'Paths analyzed normal filtered: \t\t{len(paths_normal)}')
info_util(f'Paths analyzed disintegrating: \t\t{len(paths_disintegrating)}')
info_util(f'Paths analyzed patterning defects: \t{len(paths_patterning_defects)}')
```

## Plot similarities as similarity curve<a class="anchor" id="third-bullet"></a>

```python
tools_arrow_plot(paths_normal, paths_disintegrating + paths_patterning_defects)
```

```python
tools_arrow_plot.plots(path_save=f"{dir_dst_plots}/deviation_development_predicted_stages.svg")
```
